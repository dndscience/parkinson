#METODOLOGIA CRISP-DM

<h4 align="justify">
CRISP-DM (Cross Industry Standard Process for Data Mining) é uma metodologia usada na mineração de dados para orientar o processo de descoberta de insights úteis e conhecimento a partir de dados.</h4>

<h4 align="justify">De uma forma mais simples, é uma abordagem passo a passo que ajuda-nos a extrair informações significativas ​​de grandes quantidades de dados.</h4>

<h4 align="justify">O CRISP consiste em seis etapas:

&bullet; Entendimento do negócio: definir o problema que deseja resolver e os objetivos que deseja alcançar.

&bullet; Entendimento dos dados: coletar e entender os dados que você usará para resolver o problema.

&bullet; Preparação dos Dados: limpeza e transformação dos dados para que estejam prontos para análise.

&bullet; Modelagem: construir um modelo ou modelos para ajudar a resolver o problema.

&bullet; Avaliação: avaliar a qualidade do modelo e determinar se ele atende aos requisitos do problema.

&bullet; Implementação (deployment): colocar o modelo em prática e usá-lo para resolver o problema.

<h4 align="justify">Seguindo as etapas acima, os profissionais de dados podem garantir que estão trabalhando em direção a um objetivo claro e que suas descobertas são fundamentadas em insights orientados por dados.</h4>

###Entendimento do negócio - Introdução


<h4 align="justify">Os distúrbios neurológicos como Parkinson, Alzheimer e epilepsia têm um impacto profundo na vida dos pacientes e de suas famílias. Na América do Norte, apenas a doença de Parkinson já afeta mais de um milhão de pessoas, e com a população envelhecendo, esse número deve aumentar ainda mais, devido às taxas de prevalência que tendem a crescer rapidamente após os 60 anos de idade. Além do isolamento social, há também o ônus financeiro significativo da doença, que se estima irá aumentar no futuro. Embora existam medicamentos que oferecem alívio significativo dos sintomas, principalmente nas fases iniciais, não há cura para essas condições. Como resultado, a maioria das pessoas com doença de Parkinson, por exemplo, terá uma dependência substancial da intervenção clínica.</h4>

<h4 align="justify">Para muitas pessoas com PWP, é difícil realizar as visitas físicas necessárias à clínica para monitoramento e tratamento. A expansão do acesso à internet e o aprimoramento dos sistemas de telecomunicação permitem o monitoramento remoto dos pacientes, oferecendo oportunidades significativas para reduzir o inconveniente e o custo das visitas presenciais. No entanto, para aproveitar essas oportunidades, é essencial contar com ferramentas de monitoramento clínico confiáveis.</h4>

<h4 align="justify">Estudos revelam que cerca de 90% das pessoas com PWP têm algum tipo de problema vocal. Essas alterações na voz podem ser indicadores precoces da doença e podem ser medidas de forma simples e não invasiva. Portanto, a medição da voz tem sido amplamente estudada para detectar e acompanhar a progressão dos sintomas da PD. Os sintomas vocais comuns incluem diminuição do volume, rouquidão, aspereza, redução de energia nas frequências mais altas e tremor vocal pronunciado.</h4>

<h4 align="justify">Existem vários testes vocais para avaliar os sintomas da PD. Alguns testes incluem fonações sustentadas, onde o paciente produz uma vogal e mantém o tom constante, e testes de fala contínua, onde o paciente fala uma frase padrão. Esses testes são úteis para detectar o comprometimento vocal. Além disso, existem métodos tradicionais de medição da voz, como frequência fundamental, nível de pressão sonora, jitter, shimmer e razões ruído-harmônicos. Estudos têm mostrado variações nessas medidas em pessoas com PD. Métodos inovadores baseados em sistemas dinâmicos não lineares também têm sido desenvolvidos para avaliar os sintomas disfônicos. Essas medições utilizam ferramentas de análise de séries temporais não lineares para detectar mudanças na dinâmica vocal.</h4>

<h4 align="justify">No entanto, a avaliação prática e remota da disfonia enfrenta desafios de confiabilidade devido a variações não controladas no ambiente acústico e nas características físicas dos indivíduos. Além disso, medidas tradicionais, como nível absoluto de pressão sonora, são limitadas em termos de confiabilidade em aplicações de telemedicina. Da mesma forma, a disfonia relacionada à PD é influenciada por fatores não relacionados, como preferências individuais e gênero.</h4>

<h4 align="justify">Apesar de existirem muitos métodos para avaliar distúrbios vocais e a disfonia específica da PD ser bem conhecida, ainda não há métodos eficientes para caracterizar essa disfonia considerando fatores confundidores, como o gênero do indivíduo e variações acústicas. Portanto, foi introduzida uma nova medida chamada entropia do período de pitch, que é sensível a mudanças observadas na fala relacionadas à PD.</h4>

<h4 align="justify">Relações estatisticamente significativas foram encontradas entre a disfonia na paralisia cerebral e os métodos de medição. No entanto, em condições de monitoramento remoto, as variações nas medidas são maiores do que em condições controladas. É importante avaliar a relevância prática dessas variações para determinar a gravidade da disfonia na PD. A análise discriminante linear e as máquinas de vetor de suporte são métodos preferidos para medir a capacidade de distinguir entre indivíduos saudáveis e os com PD com base nas medidas de disfonia. Esses métodos de classificação permitem combinar diferentes medidas para obter uma melhor discriminação. A taxa de classificação correta ajuda a identificar as medidas mais úteis para separar indivíduos saudáveis dos com PD em aplicações de monitoramento remoto. Além disso, é possível avaliar a eficácia dos métodos tradicionais em comparação com os novos métodos não lineares e/ou estocásticos de medição da disfonia na PD.</h4>

##Entendimento dos dados (Data understanding)





<h4 align="justify">A fase de entendimento dos dados é uma das etapas mais importantes na metodologia CRISP-DM, pois é nessa fase que o analista de dados deve explorar, analisar e compreender os dados disponíveis para posteriormente usá-los no processo de modelagem.</h4>

<h4 align="justify">Durante essa fase, o analista deve coletar informações sobre os dados, identificar possíveis problemas e entender o contexto em que eles foram gerados. É importante também verificar se os dados coletados são confiáveis, completos e representativos do problema em questão.</h4>

<h4 align="justify">Outra atividade importante nessa fase é a análise exploratória dos dados, que pode envolver a utilização de técnicas estatísticas e de visualização para identificar padrões e relações entre as variáveis. O objetivo é obter insights e entender melhor o comportamento dos dados, o que pode ajudar a orientar as próximas etapas do processo de modelagem.</h4>

###Entendimento dos dados - Projeto

<h4 align="justify">Os dados deste estudo consistem em 195 gravações de vogais sustentadas de 31 indivíduos do sexo masculino e feminino, dos quais 23 foram diagnosticados com DP. O tempo desde o diagnóstico variou de 0 a 28 anos, e as idades dos participantes variaram de 46 a 85 anos (média de 65,8, desvio padrão de 9,8). Foram registradas médias de seis gravações de vogais de cada participante, com duração variando de um a 36 segundos.</h4>

<h4 align="justify">As gravações das vogais foram feitas em uma cabine acusticamente tratada (IAC) usando um microfone de cabeça (AKG C420) posicionado a 8 cm dos lábios. O microfone foi calibrado conforme descrito, utilizando um medidor de nível sonoro de Classe 1 (B&K 2238) colocado a 30 cm do alto-falante. Os sinais de voz foram gravados diretamente no computador usando o hardware CSL 4300B (Kay Elemetrics), com taxa de amostragem de 44,1 kHz e resolução de 16 bits. Embora a normalização de amplitude afete a calibração das amostras, o estudo se concentra em medidas que não são sensíveis a alterações no nível absoluto de pressão da fala. Portanto, para garantir a confiabilidade dos algoritmos, todas as amostras foram normalizadas digitalmente em amplitude antes de calcular as medidas.</h4>



##Preparação dos dados (Data preparation)

<h4 align="justify">A fase de preparação dos dados é uma etapa crítica da metodologia CRISP-DM, pois é nessa fase que o analista deve tratar os dados brutos e prepará-los para serem usados na modelagem. Isso inclui a limpeza, transformação e seleção dos dados.</h4>

<h4 align="justify">Durante essa fase, o analista deve verificar a qualidade dos dados e corrigir quaisquer problemas identificados, como dados ausentes, inconsistentes ou errados. Também é comum a necessidade de transformar os dados, como normalização ou codificação de variáveis categóricas.</h4>

<h4 align="justify">Outra atividade importante na fase de preparação dos dados é a seleção de variáveis relevantes para o problema em questão. Isso pode envolver a análise da correlação entre as variáveis ou a utilização de técnicas de seleção de atributos. O objetivo é reduzir a dimensionalidade dos dados e melhorar a precisão dos modelos de análise de dados.</h4>

##Modelagem (Modeling)

<h4 align="justify">Na fase de modelagem da metodologia CRISP-DM, o analista de dados deve escolher a técnica de modelagem mais adequada para o problema em questão e aplicá-la aos dados preparados.</h4>

<h4 align="justify">Uma vez selecionada a técnica de modelagem, o analista de dados deve definir os parâmetros do modelo e avaliar sua precisão. Isso pode envolver a utilização de técnicas de validação cruzada ou a medição do erro de previsão do modelo.</h4>

<h4 align="justify">O objetivo final da fase de modelagem é criar um modelo que possa ser usado para prever ou classificar novos dados com um alto grau de precisão e que atenda aos objetivos do projeto.</h4>

##Avaliação (Evaluation)

<h4 align="justify">Durante a fase de avaliação da metodologia CRISP-DM, o analista de dados deve realizar uma análise crítica dos resultados obtidos na fase de modelagem. Isso envolve a verificação da qualidade dos modelos desenvolvidos e a avaliação de sua adequação aos objetivos do projeto.</h4>

<h4 align="justify">Uma das atividades mais importantes nessa fase é a validação dos modelos, que pode ser feita por meio de técnicas como validação cruzada ou divisão de conjuntos de dados. O objetivo é avaliar a precisão do modelo e determinar se ele é capaz de generalizar (ou performar) bem para novos conjuntos de dados.</h4>

<h4 align="justify">Outra atividade importante é a avaliação da utilidade do modelo para o problema em questão. Isso pode envolver a análise da relevância dos resultados produzidos pelo modelo ou a comparação com outros modelos ou métodos de resolução de problemas. O objetivo é determinar se o modelo é útil e se atende aos requisitos do projeto.</h4>

##Implementação (Deployment)

<h4 align="justify">Uma das chaves para uma implementação bem-sucedida na metodologia CRISP-DM é garantir que a solução desenvolvida esteja alinhada com os requisitos e objetivos de negócios. Por isso, é importante ter uma visão clara dos processos de negócios e das necessidades dos usuários finais antes de iniciar a implementação.</h4>

<h4 align="justify">Outro aspecto importante da fase de implementação é a necessidade de integrar a solução de dados em sistemas e processos de negócios existentes. Isso requer uma compreensão profunda dos sistemas existentes, bem como uma boa comunicação com as equipes responsáveis por esses sistemas.</h4>

<h4 align="justify">Por fim, é importante lembrar que a implementação é um processo iterativo e contínuo, e não um evento único. A solução desenvolvida deve ser monitorada e ajustada conforme necessário para garantir que continue a atender às necessidades de negócios em constante mudança.</h4>
